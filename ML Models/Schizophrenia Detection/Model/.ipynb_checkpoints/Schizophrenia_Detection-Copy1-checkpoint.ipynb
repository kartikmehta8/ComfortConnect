```python
import re
import numpy as np
import pandas as pd
import preprocess_kgptalkie as ps
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
def get_clean(x):
    x = str(x).lower().replace('\\', '').replace('_', ' ')
    x = ps.cont_exp(x)
    x = ps.remove_emails(x)
    x = ps.remove_urls(x)
    x = ps.remove_html_tags(x)
    x = ps.remove_rt(x)
    x = ps.remove_accented_chars(x)
    x = ps.remove_special_chars(x)
    x = re.sub("(.)\\1{2,}", "\\1", x)
    return x
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import LinearSVC
```

```python
tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,3), analyzer='char')
```

```python
X = tfidf.fit_transform(df['text'])
y = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
model = LinearSVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
t = "I would hear something that sound like a plane engine or like a really… you know… a really far off motor. It never went away entirely. It’s gone a lot more in the past couple of months since Christmas. It just sounds like that… it sounds like a little flame or a cellular… a digital motor"
u = "I'm going to the park to ride the wallywhoop."
```

```python
print(t)
print("-----------------------------------------------")
print(u)
```

```python
def predict_tendency(x):
    x = get_clean(x)
    vec = tfidf.transform([x])
    if model.predict(vec)[0] == 1:
        print("Schizophrenic")
    else:
        print("Not Schizophrenic")
```

```python
predict_tendency(t)
predict_tendency(u)
```

```python
model.score(x_test,y_test)
```
