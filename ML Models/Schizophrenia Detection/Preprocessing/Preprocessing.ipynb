```python
import re
import numpy as np
import pandas as pd
import preprocess_kgptalkie as ps
```

```python
df = pd.read_csv("control.csv")
```

```python
df = df.drop("text", axis=1)
```

```python
df
```

```python
def get_clean(x):
    x = str(x).lower().replace('\\', '').replace('_', ' ')
    x = ps.cont_exp(x)
    x = ps.remove_emails(x)
    x = ps.remove_urls(x)
    x = ps.remove_html_tags(x)
    x = ps.remove_rt(x)
    x = ps.remove_accented_chars(x)
    x = ps.remove_special_chars(x)
    x = re.sub("(.)\\1{2,}", "\\1", x)
    return x
```

```python
df['title'] = df['title'].apply(lambda x: get_clean(x))
```

```python
df['selftext'] = df['selftext'].apply(lambda x: get_clean(x))
```

```python
df.dropna(inplace = True)
```

```python
df.to_csv("control.csv")
```

```python
df["text"] = df["title"]+" "+df["selftext"]
```
